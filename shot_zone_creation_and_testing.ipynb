```python
import numpy as np
import pandas as pd
import random

# 1. Define the Bounding Boxes for each Zone ID
# Format: ID: [x_min, x_max, y_min, y_max]
# Scale: X=0-100 (Attacking direction), Y=0-100 (Right to Left)

zone_coordinates = {
    # --- The Box Areas ---
    3: [88, 99, 36, 64],    # Centre of the box (18-yard box, central)
    9: [88, 99, 64, 78],    # Left side of the box
    11: [88, 99, 22, 36],   # Right side of the box
    
    # --- Six Yard Box Areas ---
    10: [94, 99, 58, 64],   # Left side of the six yard box
    12: [94, 99, 36, 42],   # Right side of the six yard box
    13: [94, 99, 42, 58],   # Very close range (Central six yard box)
    
    # --- Specific Spots ---
    14: [88, 89, 49, 51],   # Penalty spot (Specific point, small variance)
    
    # --- Outside the Box ---
    15: [75, 88, 30, 70],   # Outside the box (The "D" and central area just outside)
    16: [65, 75, 25, 75],   # Long range (Further out central)
    
    # --- Wings ---
    4: [70, 99, 78, 100],   # Left wing (High up the pitch, wide left)
    5: [70, 99, 0, 22],     # Right wing (High up the pitch, wide right)
    
    # --- Difficult Angles ---
    7: [88, 99, 78, 90],    # Difficult angle on the left (Deep left)
    8: [88, 99, 10, 22],    # Difficult angle on the right (Deep right)
    6: [60, 80, 0, 100],    # Difficult angle and long range (Catch-all for bad shots)
    
    # --- Very Long Range ---
    17: [50, 65, 0, 100],   # More than 35 yards
    18: [40, 50, 0, 100],   # More than 40 yards
    
    # --- Fallback ---
    19: [0, 0, 0, 0]        # Not recorded (Will map to 0,0)
}


def generate_synthetic_coords(zone_id):
    # Get the bounds from the dictionary. If unknown, return (0,0)
    bounds = zone_coordinates.get(zone_id, [0, 0, 0, 0])
    x_min, x_max, y_min, y_max = bounds
    
    # Generate a random float within the range
    synthetic_x = random.uniform(x_min, x_max)
    synthetic_y = random.uniform(y_min, y_max)
    
    return pd.Series([synthetic_x, synthetic_y])

# --- APPLY TO YOUR DATAFRAME ---

# Assuming your DataFrame is called 'CR7' and has a 'location' column
CR7[['synth_x', 'synth_y']] = CR7['location'].apply(generate_synthetic_coords)

print(CR7[['location', 'synth_x', 'synth_y']].head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# --- 1. Define the Variables to Process ---
# Format: ("Name for File", DataFrame Variable)
# We use the variables 'Bale' and 'Benz' exactly as you have them in your environment
datasets_to_process = [
    ("Bale", Bale),
    ("Benzema", Benz)
]

# --- 2. Main Loop ---
for player_name, df in datasets_to_process:
    
    print(f"--- Processing {player_name} ---")

    # A. Create 'simplified_outcome' if it doesn't exist
    # Using the same logic: Goal, On Target (1,3), Off Target (2,4)
    if 'simplified_outcome' not in df.columns:
        df['simplified_outcome'] = df.apply(categorize_outcome, axis=1)

    # B. Create 'x' and 'y' coordinates if they don't exist
    if 'x' not in df.columns:
        coords = df['location'].apply(generate_synthetic_coords)
        df['x'] = coords[0]
        df['y'] = coords[1]
        print(f"  > Coordinates generated for {player_name}")

    # --- 3. Plotting Loop (3 Maps per Player) ---
    maps_to_generate = [
        ('Goal', '#2ecc71', f'{player_name}_heatmap_goals.png'),       # Green
        ('On Target', '#f1c40f', f'{player_name}_heatmap_ontarget.png'), # Yellow
        ('Off Target', '#e74c3c', f'{player_name}_heatmap_offtarget.png') # Red
    ]

    for outcome_name, color_code, filename in maps_to_generate:
        
        # Filter data for JUST this outcome
        subset = df[df['simplified_outcome'] == outcome_name]
        
        # Safety Check: Need at least 2 points to make a density map
        if len(subset) < 2:
            print(f"  > Skipping {outcome_name}: Not enough data points.")
            continue

        # Setup Figure (Dark Mode)
        fig, ax = plt.subplots(figsize=(12, 8))
        fig.set_facecolor('#1e1e1e')
        ax.set_facecolor('#1e1e1e')

        # Draw Pitch
        draw_pitch(ax, color='#1e1e1e', line_color='white')

        # Plot KDE
        sns.kdeplot(
            data=subset,
            x='x', 
            y='y', 
            fill=True, 
            color=color_code,
            alpha=0.6,
            levels=10,
            thresh=0.1,
            warn_singular=False,
            ax=ax
        )

        # Titles and Saving
        plt.xlim(-5, 105)
        plt.ylim(-5, 105)
        plt.title(f"{player_name} Shot Map: {outcome_name}", color='white', fontsize=18, fontweight='bold')
        plt.axis('off')
        
        plt.savefig(filename, facecolor='#1e1e1e', bbox_inches='tight')
        plt.close() # Close figure to save memory
        print(f"  > Saved {filename}")

print("\nAll Bale and Benzema maps generated successfully!")
```
