```python
import pandas as pd

# Load the events file
events = pd.read_csv("events.csv") # Removed from folder because file was too large f

# Filter to only shots
shots = events[events["event_type"] == 1].copy()

# Print total number of shots
print("Total shots:", len(shots))

# Save the shots to a new csv file
shots.to_csv("shots.csv", index=False)
```

```python
# Filter for real madrid shots
real_madrid_shots = shots[shots["event_team"] == "Real Madrid"].copy()

# Print total number of Real Madrid shots
print("Total Real Madrid shots:", len(real_madrid_shots))

# Save the Real Madrid shots to a new csv file
real_madrid_shots.to_csv("real_madrid_shots.csv", index=False)
```

```python
# Filter for Cristiano Ronaldo Shots
cristiano_shots = real_madrid_shots[real_madrid_shots["player"] == "cristiano ronaldo"].copy()

# Print total number of CR7 shots
print("Total Cristiano Ronaldo shots:", len(cristiano_shots))

# Save the CR7 shots to a new csv file
cristiano_shots.to_csv("CR7 shots.csv", index=False)
```

```python
# Filter for Karim Benzema Shots
benzema_shots = real_madrid_shots[real_madrid_shots["player"] == "karim benzema"].copy()

# Print total Number of Benzema Shots
print("Total Karim Benzema shots:", len(benzema_shots))

# Save the Benzema shots to a new csv file
benzema_shots.to_csv("Benz shots.csv", index=False)
```

```python
# Filter for Gareth Bale shots
bale_shots = real_madrid_shots[real_madrid_shots["player"] == "gareth bale"].copy()

# Print total Number of Bale Shots
print("Total Gareth Bale shots:", len(bale_shots))

# Save the Bale shots to a new csv file
bale_shots.to_csv("Bale shots.csv", index=False)
```

```python
# Load CSV with shots files for each player
CR7 = pd.read_csv("CR7 shots.csv")
Benz = pd.read_csv("Benz shots.csv")
Bale = pd.read_csv("Bale shots.csv")

# Filter for goals only
CR7_goals = CR7[CR7["is_goal"] == 1]
Benz_goals = Benz[Benz["is_goal"] == 1]
Bale_goals = Bale[Bale["is_goal"] == 1]

# Save to CSV
CR7_goals.to_csv("CR7 goals.csv", index=False)
Benz_goals.to_csv("Benz goals.csv", index=False)
Bale_goals.to_csv("Bale goals.csv", index=False)

# Print Total Goals per Player
print("Total CR7 Goals:", len(CR7_goals))
print("Total Benz Goals:", len(Benz_goals))
print("Total Bale Goals:", len(Bale_goals))
```
